```python
i = 0
f = open("assessments.csv", "r")
for line in f:
    cols = line.split(",")
    print cols[3], "\t", cols[34]
    i = i + 1 # or i+=1
    if i >= 10:
        break
f.close()
```

```python
i = 0
with open("assessments.csv", "r") as f: #convenient for doing file processing
    for line in f:
        cols = line.split(",")
        print cols[3], "\t", cols[34]
        i = i + 1 # or i+=1
        if i >= 10:
            break
```
